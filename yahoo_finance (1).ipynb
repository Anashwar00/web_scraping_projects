```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url='https://finance.yahoo.com/gainers'
```

```python
page=requests.get(url)
```

```python
soup=bs(page.text,'html')
```

```python
soup
```

```python
table=soup.find('table',class_="W(100%)")
```

```python
data_set=table.find('tr')
lst=[]
```

```python
for data in data_set:
    lst.append(data.text.strip())
#     for i in data:
#         print(i)
#         print(i.text.strip())
    
```

```python
lst
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=lst)
```

```python
df
```

```python
#retreving row data
```

```python
dat_set=table.find_all('tr',class_="simpTblRow")
len(dat_set)
```

```python
for data in dat_set:
    ind_row=data.find_all('td')
#     print(ind_row)
    final_row=[data.text.strip()for data in ind_row]
#     leght=len(df)
    df.loc[len(df)]=final_row
df
```
