# Web_Scraping_Ebay


```python
# import libraries
from bs4 import BeautifulSoup as bs
import requests
```

```python
#requesting server for url
url='https://www.ebay.com/sch/i.html?_from=R40&_nkw=toy+car&_sacat=0&_pgn=35'
page=requests.get(url)
```

```python
#getting website code in text formet
soup=bs(page.text,'html')
```

```python
soup
```

```python
#Finding ul so that we can find desired result which is present inside ul tag
table_stuff=soup.find('ul',class_="srp-results srp-grid clearfix")
```

```python
b=table_stuff.find_all('span',class_="s-item__price")
```

```python
#list comprehension
price=[i.text.strip()for i in b]
```

```python
a=table_stuff.find_all('div',class_="s-item__title")
a
```

```python
descp=[i.text.strip()for i in a]
descp
```

```python
#combing lists
combined_data = list(zip(descp, price))

print(combined_data)
```

```python
import pandas as pd
# making data frame
df = pd.DataFrame(combined_data, columns=['descp', 'price'])

df
```
